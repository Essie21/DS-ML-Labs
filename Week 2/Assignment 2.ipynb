# Operators

### 1. Arithmetic Operators

```python
# Addition (+)
x = 20
y = 4
print (x + y)
```

```python
# Subtraction (-)
print (x - y)
```

```python
# Multiplication (*) - An asterisk is used
print (x * y)
```

```python
# Division (/) - A forward slash is used
print (x / y)
```

```python
# Floor Division (//) - When this is used, should the quotient come with a whole number then decimals, the drcimals are ignored and just the whole number is used.
print (x // y)
```

```python
# Modulo (%) - The modulo is the opposite of the floor division. It gives only the remainder left after the division is done)
print (x % y)
```

```python
# Exponential (**) - Double asterisks are used. It makes the second variable the power of the first variable
x = 5
y = 3
print (x ** y)
```

### 3. Logical Operators

```python
# AND (Both conditions have to be true for the output to be true, output is false if one condition is false)
x = 32
y = 21
print (x > 30 and y < 25)
```

```python
# OR (If any of the conditions is true, the output will be true)
print (x < 30 or y < 25) 
# Output is true because y is less than 25 even though x is not less than 30)
```

```python
# NOT (Usually negates the condition set)
not (x < 32)
# The condition is supposed to be false because x is equal to 32 not less than. The "not" operator negates the false output value making it true. 
```

### 5. Bitwise Operations
It references integers in binary levels (0's and 1's). The output has to be in binary figures.

6 Operators to be used: ( &, |, ^, ~, <<, >> )

```python
# AND (&)
x = 55 # Binary: 00110111
y = 70  # Binary: 01000110

# 55: 0 0 1 1 0 1 1 1
# 70: 0 1 0 0 0 1 1 0
# --- -----------------
result = x & y 
# 55 & 70: 0 0 0 0 0 1 1 0  = 6
# Output is 6
print (result)
```

```python
# OR (|) 
x = 55 # Binary: 00110111
y = 70  # Binary: 01000110

# 55: 0 0 1 1 0 1 1 1
# 70: 0 1 0 0 0 1 1 0
# --- -----------------
result = x | y 
# 55 | 70: 0 1 1 1 0 1 1 1  = 119
# Output is 119
print (result)
```

```python
#Exclusive OR (^) 
x = 55 # Binary: 00110111
y = 70  # Binary: 01000110

# 55: 0 0 1 1 0 1 1 1
# 70: 0 1 0 0 0 1 1 0
# --- -----------------
result = x ^ y 
# 55 ^ 70:  0 1 1 1 0 0 0 1  = 113
# Output is 113
print (result)
```

```python
# NOT Operator (~) - This negates (flips) all the binary digits of the value
x = 55 # Binary: 00110111
result = ~x
# ~55 : 1 1 0 0 1 0 0 0 = -56
# Output is -56
print (result)
```

```python
# Left Shift (<<) - This operator shift the bits of a number to the left by a specified number of positions.
x = 55 # Binary: 00110111
result = x << 2
# Left shift adds zeros to the right and effectively multiplies by 2ⁿ.

#55:     0 0 1 1 0 1 1 1
#        ← ← ← ← ← ← ← ←  (shift to left by 2) or remove first two digits on the left. 
#        55 << 2:   1 1 0 1 1 1 0 0  (add 2 zeros on the right) = 220
# Output is 220
print (result)
```

```python
# Right Shift (>>) - This operator shift the bits of a number to the right by a specified number of positions.
x = 55 # Binary: 00110111
result = x >> 2
# Right shift adds zeros to the left and effectively multiplies by 2ⁿ.

#55:     0 0 1 1 0 1 1 1
#        → → → → → → → →  (shift to right by 2) or remove last 2 digits on the right
#        55 >> 2:   0 0 0 0 1 1 0 1 (add 2 zeros on the left) = 13
# Output is 13
print (result)
```

### 6. Membership Operators
They check whether a value exists in a sequence or collection (strings, lists, tuples, sets, dictionaries)

Has 2 operators: (in, not in). Output is a boolean

```python
a = [1, 2, 3, 4, 5]
```

```python
# in - Checks whether a value exist in a sequence
print ( 2 in a)
# There 2 in the list a, so the output is true
```

```python
print (8 in a)
# 8 is not in the list a so the output is false
```

### 7. Identity Operators
They check if two objects are the same objects in memory with equal values. Compares the memory occupied by the values not the values themselves 

Has 2 operators: (is, is not)

```python
a = [1, 2, 3, 4, 5]
b = [1, 2, 3, 4, 5]

print ( a is b)
# The output is false since the memory occupied by a is not the same as the memory occupied by b
```

```python
print (a is not b)
# The output is true since the memory occupied by a is not the same as the memory occupied by b
```
