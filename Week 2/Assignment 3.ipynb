# Strings and Control Flow

#### Operations on Strings

```python
name_ = "ParoCyber LLC"
print (name_)
```

```python
# Changing all letters to uppercase
# upper ()
name_= name_.upper()
print (name_)
```

```python
# Changing all letters to lowercase
# lower ()
name_= name_.lower()
print (name_)
```

```python
# isalpha () - Checks whether all the characters present are alphabets
name_= name_.isalpha()
print (name_)
```

```python
# isdigit () - Checks whether all the characters present are digits (numbers)
Name_= "ParoCyber LLC"
result = Name_.isdigit()
print (result)
# Output is false because there is no number
```

```python
#islower() - Checks whether all the characters present are in lower case
result = Name_.islower()
print (result)
# Output is false because not all the letters are in lowercase
```

```python
#isupper() - Checks whether all the characters present are in upper case
result = Name_.isupper()
print (result)
# Output is false because not all the letters are in uppercase
```

#### How to split strings

```python
# split () - Consider the character that separates each word
Flowers = "Last week couples got flowers" # (Separated by space)
Flowers.split(" ") 

# Separates the word by the spaces between them
```

#### How to join strings

```python
# join ()
Flowers = "Last week couples got flowers" 
word = Flowers.split (" ")
print ("_".join(word))
# This outputs the words in flowers separated by "_"
```

```python
Flowers = "Last week couples got flowers"
words = Flowers.split(" ")  # Split into list: ['Last', 'week', 'couples', 'got', 'flowers']

# Join them back with a space
result = " ".join(words)
print(result)  # Output: Last week couples got flowers
```

```python
# replace - Replaces a character(s) with a new one
Flowers = "Last week couples got flowers"
word = Flowers.replace ("flowers", "chocolate")
print (word)
# This replaces "flowers" with "chocolate"
```

```python
# startswith - To see what an element/string starts with
Flower = "roses"
Flower.startswith("s")
# The output will be false because roses does not start with an "s"
```

```python
Flower = "roses"
Flower.startswith("r")
# The output is true since roses starts with an "r"
```

```python
# endswith: To see what an element/string ends with
Flower = "Daisy"
Flower.endswith("e")
# The output will be false since Daisy does not end with an "e"
```

```python
Flower = "Daisy"
Flower.endswith("y")
# The output will be true since Daisy ends with a "y"
```
