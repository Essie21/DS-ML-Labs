## Lists

#### Lists are ordered collections of items in Python. They are mutable, meaning their elements can be modified after creation.
#### Lists are created using square brackets [] or the list() constructor.

```python
#Creating a list
letters = ["a", "b", "c", "d", "e"]
letters
```

```python
#Append: Adds an element to the end of the list
letters.append("f")
print(letters)
```

```python
#Extend: Extends the list by appending elements from another list.
letters.extend(["g", "h"])
letters
```

```python
#Insert: Inserts an element at a specified position.
letters.insert(0,1)
letters
```

```python
# Remove: Removes the first occurrence of a specified value.
letters.remove("b")
letters
```

```python
# Pop: Removes and returns the element at a specified position (default is the last element
pop_element = letters.pop(5)
print(pop_element)
print(letters)
```

```python
#Index: Returns the position of the letter "g"
Index = letters.index("g")
print(Index)
```

```python
#Count: Returns the number of times a value occurred 
Count = letters.count("d")
print(Count)
```

```python
#Sorting: Sorts in ascending order
letters.sort()
letters
```

```python
#Reverse: Sorts in descending order
letters.reverse()
letters
```

## Tuples
Their contents can't be changed. Constructed using normal brackets ()

```python
countries = ("Ghana", "India", "China", "Togo", "Benin")
countries
```

```python
#Count: Returns the number of times a value occurred
count = countries.count("Benin")
print(count)
```

```python
#Index: Returns the position of the country "China"
index = countries.index("China")
print(index)
```

## Dictionaries
An unordered collection for key-value pairs. 
Can be created in these formats: 
1. {"key": "value"} or
2. dict(["key", "value"])
3. dict(key = "value")

```python
#Creating a dictionary - First Way (Using Curly brackets)
myDictionary = {"Name": "Eric", "Age": 21, "Hobby": "Painting"}
myDictionary
```

```python
#Creating a dictionary - Second Way (Using keyword "dict" with normal brackets)
myDictionary = dict([("Name", "Eric"), ("Age", 21), ("Hobby", "Painting")])
myDictionary
```

```python
#Creating a dictionary - Third Way (Using keyword "dict" with = and normal brackets)
myDictionary = dict(Name = "Eric", Age = 21, Hobby = "Painting")
myDictionary
```

```python
#Creating Dictionary with a key having multiple values (Using Way 1) - Best Way
myDictionary = {"Name": ("Eric", "Cynthia", "Mary"), "Age": (21, 34, 45), "Hobby": ("Painting", "Singing", "Dancing")}
myDictionary
```

```python
#Creating Dictionary with a key having multiple values (Using Way 2)
myDictionary = dict([("Name", ["Eric", "Cynthia", "Mary"]), ("Age", [21, 34,45]), ("Hobby", ["Painting", "Singing", "Dancing"])])
myDictionary
```

```python
#Getting the age
age = myDictionary.get("Age")
print(age)
```

```python
#Getting Items: This Returns all the items in the dictionary
Items = myDictionary.items()
Items
```

```python
#Getting keys: Returns all keys in the dictionary 
Keys = myDictionary.keys()
Keys
```

```python
#Getting Values:
Values = myDictionary.values()
Values
```

```python
#Updating the dictionary by adding a key named gender

myDictionary.update ({"Gender": ["Male", "Female", "Female"]})
myDictionary
```
